```python
import pykitti
import numpy as np

data = pykitti.raw(
    base_path=r"C:\Users\disha\Project\datasets\kitti",
    date='2011_09_26',
    drive='0009'
)

# Available timestamps in this PyKITTI version
timestamps = np.array([t.timestamp() for t in data.timestamps])

print(f"Total synchronized timestamps: {len(timestamps)}")
print(f"GPS/IMU frames: {len(data.oxts)}")
print(f"LiDAR frames:   {len(data.velo_files)}")
print(f"Camera frames:  {len(data.cam2_files)}")

# Frame count differences
print("\nFrame count differences:")
print(f"GPS/IMU - LiDAR  = {len(data.oxts) - len(data.velo_files)}")
print(f"GPS/IMU - Camera = {len(data.oxts) - len(data.cam2_files)}")

# Check timestamp spacing
gaps = np.diff(timestamps)

print(f"\nMean frame interval: {gaps.mean()*1000:.2f} ms")
print(f"Min frame interval:  {gaps.min()*1000:.2f} ms")
print(f"Max frame interval:  {gaps.max()*1000:.2f} ms")
```
