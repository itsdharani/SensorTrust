```python
import pykitti
import numpy as np

# Load sequence 0009
data = pykitti.raw(
    base_path=r"C:\Users\disha\Project\datasets\kitti",
    date='2011_09_26',
    drive='0009'
)

# Check GPS/IMU loaded
print(f"Number of GPS/IMU frames: {len(data.oxts)}")

# Check LiDAR loaded
print(f"Number of LiDAR scans: {len(data.velo_files)}")

# Check camera loaded
print(f"Number of camera frames: {len(data.cam2_files)}")
```

```python
# Get the first GPS/IMU reading
frame = data.oxts[0]

print(f"Latitude:          {frame.packet.lat}")
print(f"Longitude:         {frame.packet.lon}")
print(f"Forward velocity:  {frame.packet.vf} m/s")
print(f"Acceleration x:    {frame.packet.ax} m/s²")
print(f"Acceleration y:    {frame.packet.ay} m/s²")
print(f"Acceleration z:    {frame.packet.az} m/s²")
```

```python
import open3d as o3d

# Load the first scan
scan = data.get_velo(0)   # returns (N, 4) array: x, y, z, intensity
print(f"Number of points in scan: {scan.shape[0]}")
print(f"First 5 points:\n{scan[:5]}")

# Visualise it (opens a 3D window — close it when done)
points = scan[:, :3]
pcd = o3d.geometry.PointCloud()
pcd.points = o3d.utility.Vector3dVector(points)
o3d.visualization.draw_geometries([pcd])
```
